```python
import matplotlib.pyplot as plt
import numpy as np

low_alb = np.loadtxt("../Albedo_runs/low_albedo_test/snow_status.dat",delimiter=",")
high_alb = np.loadtxt("../Albedo_runs/high_albedo_test/snow_status.dat",delimiter=",")
int_alb = np.loadtxt("../Albedo_runs/Albedo_baseline/snow_status.dat",delimiter=",")


time = int_alb[:,0]/86400
sd_int = int_alb[:,1]
sd_low = low_alb[:,1]
sd_high = high_alb[:,1]

fig1 = plt.figure(figsize=(10,5))
plt.plot(time,sd_int,color='k', label="internal albedo")
plt.plot(time,sd_low,color='b', label="albeo = 0.3")
#plt.plot(time,sd_high,color='r', label="albeo = 0.9")
plt.legend(frameon=False)
plt.xlabel("time (days)")
plt.ylabel("snow depth (m)")
plt.show()
```

```python
fig2 = plt.figure(figsize=(5,5))
plt.plot(time,sd_int*100.0,color='k', label="internal albedo")
plt.plot(time,sd_low*100.0,color='b', label="albeo = 0.3")
#plt.plot(time,sd_high,color='r', label="albeo = 0.9")
plt.xlim([1450,1600])
plt.legend(frameon=False)
plt.xlabel("time (days)")
plt.ylabel("snow depth (cm)")
plt.show()
```
